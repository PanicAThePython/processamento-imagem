```python
# Natália Sens Weise e Matheus Petters Bevilaqua
from matplotlib import pyplot as plt
import cv2
import numpy as np

img = cv2.imread("Cagado.jpg")
## links de referencia: dontpad.com/pecagadoado

kernel = np.ones((4, 4),np.uint8)

# aplicando filtros
# grayscale (auxilia ao transformara pra preto e branco)
grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# preto e branco (black and white -> bnw)
(thresh, bnw) = cv2.threshold(grayscale, 125, 255, cv2.THRESH_BINARY)
# aplicando morph de abertura (erosão+dilatação)
openingbnw = cv2.morphologyEx(bnw, cv2.MORPH_OPEN, kernel)
# invertendo a binarização
reverse = cv2.bitwise_not(openingbnw)
#erodindo
# duas iterações pq fica vazia o quadrante inferior esquerdo ficar mais parecido com a meta, mas bom rever talvez, ver a opinião do prof
erosion = cv2.erode(reverse,kernel,iterations = 2)

plt.imshow(img, 'gray')
plt.title('imagem')
plt.show()
plt.imshow(grayscale, 'gray')
plt.title('filtro cinza')
plt.show()
plt.imshow(bnw, 'gray')
plt.title('preto e branco')
plt.show()
plt.imshow(openingbnw, 'gray')
plt.title('abertura')
plt.show()
plt.imshow(reverse, 'gray')
plt.title('inversão')
plt.show()
plt.imshow(erosion, 'gray')
plt.title('erosão - resultado final')
plt.show()
```

```python
# Natália Sens Weise e Matheus Petters Bevilaqua
from matplotlib import pyplot as plt
import cv2
import numpy as np

img = cv2.imread("Quadra.jpg")
kernel = np.ones((3, 3),np.uint8)

plt.imshow(img, 'gray')
plt.title('imagem')
plt.show()
## links de referencia: dontpad.com/pecagadoado

grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grayscale, 'gray')
plt.title('filtro cinza')
plt.show()
(thresh, bnw) = cv2.threshold(grayscale, 80, 255, cv2.THRESH_BINARY)

# fechamento
closing = cv2.morphologyEx(bnw, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing, 'gray')
plt.title('fechamento')
plt.show()

# abertura
opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel)
plt.imshow(opening, 'gray')
plt.title('abertura')
plt.show()

# erosão
erosion = cv2.erode(opening,kernel,iterations = 1)
plt.imshow(erosion, 'gray')
plt.title('erosão')
plt.show()

# abertura
opening2 = cv2.morphologyEx(erosion, cv2.MORPH_OPEN, kernel)
plt.imshow(opening2, 'gray')
plt.title('abertura - resultado final')
plt.show()
```
