```python
# Natália Sens Weise e Matheus Petters Bevilaqua
import numpy as np
import cv2
from matplotlib import pyplot as plt

# selecionando imagens
img1 = cv2.imread('pare.jpg',cv2.COLOR_BGR2GRAY)          # query image
img2 = cv2.imread('imagemPare_11.jpg',cv2.COLOR_BGR2GRAY) # train image

# iniciando o ORB
orb = cv2.ORB_create()

# achando os keypoints e descritores
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)

# criando BFMatcher
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1,des2)

# Desenhando os pontos.
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches,None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
plt.imshow(img3),plt.show()

#tabela ORB

# imagem              -----  resultado
# imagem50km_01         | Maioria esmagadora dos pontos fora
# imagem50km_02         | Uns 5 ou 6 dentro, mas o resto tudo fora (muitos)
# imagem50km_03         | Todos os pontos a excessão de um fora
# imagem50km_04         | Uma boa parte dentro, mas ainda tinha varios fora
# imagem50km_05         | Cerca de metade dos pontos dentro
# imagem50km_06         | Quase metade dos pontos dentro
# imagem50km_07         | Varios pontos dentro, mas vários fora também. Os de dentro bem focados e os de fora mais dispersos, embora ainda proximos a placa
# imagem50km_08         | Um ou outro dentro, mas a maioria fora, dispersos pelos carros
# imagem50km_09         | Quase todos os pontos dentro, só uns 3 fora.
# imagem50km_10         | Alguns na placa, mas a maioria dispersos pela traseira da kombi
# imagem50km_11         | Nenhum ponto batendo na placa de 50km. Alguns poucos pontos batendo no papel em baixo e na placa de "caminhao, mantenha-se a direita" mas a grande maioria fora
#
# imagemLombada_01      | Poucos pontos reconhecidos, todos fora
# imagemLombada_02      | Poucos pontos no total novamente, nenhum na placa, alguns mais focados na iluminação traseira do carro
# imagemLombada_03      | Poucos pontos, nenhum na placa. Focos maiores na iluminação traseira do carro e na base de uma árvore
# imagemLombada_04      | Pontos bem dispersos, um deles na placa de pedestres atrás, mas nenhum na de lombada
# imagemLombada_05      | Pontos dispersos, novamente nenhum na imagem
# imagemLombada_06      | Quatro pontos batendo na placa, nas bordas, e alguns outros espalhados em alguns focos aleatórios.
# imagemLombada_07      | Um ponto na placa, e varios na paisagem na esquerda da imagem
# imagemLombada_08      | Pontos bem dispersos, pegando em alguins carros mas em nenhuma das placas de lombada, (e um só nas placas em geral)
# imagemLombada_09      | Dois ou três pontos na borda da placa, o resto tudo batendo na folhagem.
# imagemLombada_10      | Pontos bem dispersos, varios pequenos focos mas nenhum na placa
# imagemLombada_11      | Alguns pontos na placa de trepidação, mas nenhum na de lombada
#
# imagemPare_01         | Varios pontos na placa, mas com mais uns 3 ou 4 focos tão numerosos quanto. Alguns pontos batendo na marca dágua
# imagemPare_02         | Pontos bem dispersos, a maioria esmagadora na folhagem. Ou ou outro em cada placa de pare
# imagemPare_03         | Grande foco no centro da imagem, com quase todos os pontos. Nenhum na placa.
# imagemPare_04         | Alguns pontos na placa, mas a maioria fora.
# imagemPare_05         | Alguns pontos na placa, mas a maioria fora, com focos maiores onde tem escritos (muro/fachada no fundo)
# imagemPare_06         | A grande maioria na placa, mas alguns poucos na placa de trás
# imagemPare_07         | Pontos espalhados em alguns focos principais (roupas dos garis, placa de pedestre) mas alguns poucos na placa de pare. Varios solitarios espalhados
# imagemPare_08         | Um ou dois pontos na placa. Do resto a maioria nos carros ou no ambiente
# imagemPare_09         | Nenhum ponto na placa em si, mas notam-se vários no "Pare" escrito no chão
# imagemPare_10         | Cerca de 1/5 dos pontos na placa da esquerda, mas nenhum na da direita. Alguns pontos na outras placas e outros dispersos pelo ambiente
# imagemPare_11         | Poucos pontos na placa, nenhum no escrito do chão. Quantidade assustadora de pontos batendo na folhagem


```

```python
### Parte 2, stitching
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("img.PNG",cv2.COLOR_BGR2GRAY)
img = cv2.resize(img,(0,0),fx=0.4,fy=0.4)
img1 = cv2.imread("img1.PNG",cv2.COLOR_BGR2GRAY)
img1 = cv2.resize(img1,(0,0),fx=0.4,fy=0.4)
img2 = cv2.imread("img2.PNG",cv2.COLOR_BGR2GRAY)
img2 = cv2.resize(img2,(0,0),fx=0.4,fy=0.4)
# a segunda linha em cada imagem serve pra dar resize em todas elas pra um tamanho comum
# sem o resize dava mto bug na hora do stitch, e agora fica mais consistente

imgs = []
imgs.append(img)
imgs.append(img1)
imgs.append(img2)

# mostrando as 3 imgs
# plt.imshow(img),plt.show()
# plt.imshow(img1),plt.show()
# plt.imshow(img2),plt.show()

# mode pode ser 0 (panorama) ou 1 (scanner)
# o panorama distorcia muito o topo da imagem, então vamos de scanner, que distorce menos
stitchy=cv2.Stitcher.create(mode=1)
(status,output)=stitchy.stitch(imgs)

plt.imshow(output)
```
